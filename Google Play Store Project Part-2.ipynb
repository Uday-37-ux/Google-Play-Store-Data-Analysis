```python
import pandas as pd
pd.set_option("display.max_columns",None)
```

```python
df=pd.read_csv("D:\\projects\\googleplaystore.csv")
print(df)
```

```python
#9. Which Category Getting The Highest Average Rating?
df.groupby("Category")["Rating"].mean().sort_values(ascending=False).head(1)
```

```python
#OR
#9. Which Category Getting The Highest Average Rating?
df.groupby("Category")["Rating"].mean().sort_values().tail(1)
```

```python
#10.Find the total number App having 5 star Rating Apps
star5apps=len(df[df["Rating"]==5])
print("total number App having 5 star Rating Apps=",star5apps)
```

```python
#Display the Apps with 5 star Rating
for appname in df[df["Rating"]==5].iterrows():
    print(appname)
```

```python
#11.Find the Average Values of Reviews
df["Reviews"]=df["Reviews"].replace("3.0M",3.0*1000000)
df["Reviews"]=df["Reviews"].astype("float")
mv=df["Reviews"].mean()
print("the Average Values of Reviews=",mv)
```

```python
#12.Find the total number of Free and Paid Apps of Google Play store
df["Type"].value_counts()
```

```python
#13. Find which app has Maximum Review
appname=df[df["Reviews"].max()==df["Reviews"]]["App"]
print("The app has Maximum Review=",appname)
```

```python
#14. Find the Top 5 App Names having highest Review
df["Reviews"].sort_values(ascending=False).head()
```

```python
indices=df["Reviews"].sort_values(ascending=False).head().index
df.iloc[indices]["App"]
```

```python
df.iloc[indices]["App"].unique()
```

```python
#15.Find Average Rating of Free and Paid Apps
df.groupby("Type")["Rating"].mean()
df.groupby("Type")["Rating"].mean()
```

```python
#16. Find Top 5 Apps Having Maximum Installs
df["Installs_1"]=df["Installs"].str.replace(",","")
df['Installs_1']=df['Installs_1'].str.replace("+","")
df['Installs_1']
```

```python
df["Installs_1"]=df["Installs_1"].str.replace("Free","0")
df["Installs_1"]=df["Installs_1"].astype("int")
df["Installs_1"].dtype
```

```python
indices=df["Installs_1"].sort_values(ascending=False).head().index
df.loc[indices]["App"]
```
